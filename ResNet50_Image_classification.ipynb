```
import numpy as np
from keras.preprocessing import image
from keras.applications import resnet50
```

```
model = resnet50.ResNet50()
```

# Code to import any type of file in google colab

```
from google.colab import files
from io import BytesIO
from PIL import Image

uploaded = files.upload()
```

#Display image in google colab after importing

```
im = Image.open(BytesIO(uploaded['bay.jpg']))
import matplotlib.pyplot as plt
plt.imshow(im)
plt.show()
```

#Resizing the image for ResNet50

```
img = image.load_img("bay.jpg",target_size=(224,224))
```

# Next we will convert the image to numbers to feed it to the neural network i.e. into a numpy array

```
x = image.img_to_array(img)
x
```

#add another dimension to input a list of images, rather than just one

```
x=np.expand_dims(x,axis=0)
```

#preprocessing the images to scale them as numpy allows the image size upto 255 bits that is each pixel occupies upto one byte. Calling an in-built function for this

```
x=resnet50.preprocess_input(x)
```

#Run the image through a deep neural network to make a prediction. The prediction object is an array with 1000 points and at each point we get the prediction of the liklihood of our input image vs data on which resnet50 is trained

```
predictions=model.predict
```

```
predicted_class=resnet50.decode_predictions(predictions,top=20)
```

```
print("This is an image of:")

for imagenet_id, name, likelihood in predicted_class[0]:
    print(" - {}: {:2f} likelihood".format(name, likelihood))
```
